```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import pymysql
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20221113'

res = requests.get(url)
soup = bs(res.text, 'html.parser')
```

```python
result = soup.select('.list_ranking tr')
```

```python
for i in result:
    if i.select_one('.tit5 a') != None:
        print(i.select_one('.tit5 a').text)
    if i.select_one('.point') != None:
        print(i.select_one('.point').text)
    if i.select_one('.range') != None:
        print(i.select_one('.range').text)
    if i.select('img') != []:
        print(i.select('img')[2].attrs['alt'])
    
    print('-'*80)
    
        
        
```

```python
db = pymysql.connect(host = 'localhost',port=3306,
                    user = 'root', passwd = 'dbwls4343*',
                    db = 'pydb', charset = 'utf8')
```

```python
cursor = db.cursor()
```

```python
sql = '''
CREATE TABLE movie_grade(
    title VARCHAR(100),
    score VARCHAR(10),
    fluct VARCHAR(10),
    updown VARCHAR(10)
    )'''
```

```python
cursor.execute(sql)
db.commit()
```

```python
for i in result:

    if i.select_one('.tit5 a') == None:
        continue

    if i.select_one('.tit5 a') != None:
        # print(i.select_one('.tit5 a').text)
        title = i.select_one('.tit5 a').text

    if i.select_one('.point') != None:
        # print(i.select_one('.point').text)
        score = i.select_one('.point').text

    if i.select_one('.range') != None:
        # print(i.select_one('.range').text)
        fluct = i.select_one('.range').text

    if i.select('img') != []:
        # print(i.select('img')[2].attrs['alt'])
        updown = i.select('img')[2].attrs['alt']

    sql = f'''
        INSERT INTO movie_grade
        VALUES ('{title}', '{score}', '{fluct}', '{updown}')
    '''
    cursor.execute(sql)
    db.commit()

    print(f'-->지금 영화 [{title}] 정보가 DB에 저장됐습니다.')
    print('-' * 50)
    print()
```

```python
sql = '''
SELECT * 
FROM movie_grade
'''
cursor.execute(sql)
rs = cursor.fetchall()
df = pd.DataFrame(data = rs,
                columns = ['title', 'score','fluct','updown'])
df
```
