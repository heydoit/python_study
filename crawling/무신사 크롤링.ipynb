```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://www.musinsa.com/categories/item/001"
```

```python
res = requests.get(url)
```

```python
soup = bs(res.text, "html.parser")
```

```python
result = soup.select('#searchList > li')
```

```python
if len(soup.select_one('.price').text.strip().split()) == 1:
    print('Price : ', soup.select_one('.price').text.strip().split()[0])
else:
    print('Price : ', soup.select_one('.price').text.strip().split()[1])
```

```python
soup.select_one('.price').text.strip().split()
```

```python
for i in result:
    print('Brand : ', i.select_one('.item_title a').text.strip())
    print('Name : ', i.select_one('.list_info a').attrs['title'].strip())
    if len(i.select_one('.price').text.strip().split()) == 1:
        print('Price : ', i.select_one('.price').text.strip().split()[0])
    else:
        print('Price : ', i.select_one('.price').text.strip().split()[1])
    print('Buy_count : ', i.select_one('.count').text.strip(),'개')
    
    print('---------')
```
