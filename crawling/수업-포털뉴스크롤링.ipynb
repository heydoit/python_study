```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid2=230&sid1=105&date=20221120"
```

```python
headers = ({
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
})
```

```python
res = requests.get(url, headers = headers)
soup = bs(res.text, 'html.parser')
```

```python
result = soup.select('.list_body ul li')
```

```python
for i in result:
    title = i['alt'].strip()
    link = i['src'].strip()
    print(title, link)
```

```python
for count, i in enumerate(result):
    news_url = i.select('dt')[1].select_one('a').attrs['href']
    news_reaponse = requests.get(news_url, headers = headers)
    news_soup = bs(news_reaponse.text, 'html.parser')
    news_title = news_soup.select_one('div.media_end_head_title h2.media_end_head_headline').text.strip()
    news_content = news_soup.select_one('div#newsct_article').text.strip()
    news_date = news_soup.select_one('div.media_end_head_info_datestamp_bunch span.media_end_head_info_datestamp_time').text.strip()
    print(news_title)
    print(news_content)
    print(news_date)
    print('-'*80)
```
