```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://play.google.com/store/movies/collection/topselling_paid?clp=ChcKFQoPdG9wc2VsbGluZ19wYWlkEAcYBA%3D%3D:S:ANO1ljLCcE4&gsr=ChkKFwoVCg90b3BzZWxsaW5nX3BhaWQQBxgE:S:ANO1ljKn82s&hl=ko&gl=US'
res = requests.get(url)
soup = bs(res.text, 'html.parser')
```

```python
result = soup.select('div.fUEl2e.rorPIe a.Si6A0c.ZD8Cqc')
result
```

```python
count = 0
for i in result:
        print(i)
        print('-'*70)
        count += 1
```

```python
index = 1
for i in result:
    print(index,'영화제목 : ', i.select_one('div.Epkrse').text)
    print('평점 : ', i.select_one('div.LrNMN').text[:3],'점')
    print(i.select_one('span.VfPpfd.VixbEe').text)
    print('-'*70)
    index +=1
```

```python
import os
fd = 'googleplay_mv'

if os.path.isdir(fd):
    print(f'폴더 {fd}가 이미 작업경로에 존재합니다.')
else:
    os.mkdir(fd)
    print(f'폴더 {fd}를 현재 작업경로에 생성했습니다.')
```

```python
items = []

for i in result:
    mv_title = i.select_one('div.Epkrse').text
    mv_score = i.select_one('div.LrNMN').text[:3]
    mv_price = i.select_one('span.VfPpfd.VixbEe').text
    
    items.append([mv_title, mv_score, mv_price])
    
    img_url = i.select_one('div.TjRVLb > img')['src']
    img_res = requests.get(img_url)
    img_title = i.select_one('div.TjRVLb > img')['alt']

    with open(f'./googleplay_mv/{img_title}.jpg', mode ='wb') as f:
        f.write(img_res.content)
```

```python
df = pd.DataFrame(items, columns = ['Title','Score','Price'])
df
```

```python
browser = webdriver.Chrome()
browser.get(url)


prev_height = browser.execute_script('return document.body.scrollHeight')
interval = 3
count = 0

while True:
    browser.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    

    print(f'스크롤 하단이동 작업을 {count + 1}회 수행했습니다.')
    print('-'*80)
    time.sleep(interval)
    
    curr_height = browser.execute_script('return document.body.scrollHeight')
    
    if curr_height == prev_height:
        break
    
    prev_height = curr_height
    
    count += 1
    
print('#'*80)
print('스크롤 하단이동 작업이 완료됐습니다.')
print('#'*80)

items = []

for i in enumerate(result):
    mv_title = i.select_one('div.Epkrse').text
    mv_score = i.select_one('div.LrNMN').text[:3]
    mv_price = i.select_one('span.VfPpfd.VixbEe').text
    
    items.append([mv_title, mv_score, mv_price])
    
    img_url = i.select_one('div.TjRVLb > img')['src']
    img_res = requests.get(img_url)
    img_title = i.select_one('div.TjRVLb > img')['alt']

    with open(f'./googleplay_mv/{img_title}.jpg', mode ='wb') as f:
        f.write(img_res.content)
        
df = pd.DataFrame(items, columns = ['Title','Score','Price'])
df
```
