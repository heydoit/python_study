```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20221123"
```

```python
res = requests.get(url)
```

```python
soup = bs(res.text, "html.parser")
```

```python
movies = soup.select_one(".list_ranking tr a")
```

```python
movies
```

```python
result = soup.select('.list_ranking tr')
```

```python
result
```

```python
movie_grade=[]
for i in result:
    if i.select_one('div a') != None:
        #print("Movie Title : ", i.select_one('div a').text)
        mv_title = i.select_one('div a').text
    if i.select_one('.point') != None:
        #print("Mark : ", i.select_one('.point').text)
        mv_mark = i.select_one('.point').text
    if i.select_one('.range') != None:
        #print("Ranking Change : ", i.select_one('.range').text)
        mv_ranking = i.select_one('.range').text
    if i.select_one('.ac') != None:
        #print("Up-Down : ", i.select('img')[2].attrs['alt'])
        mv_ud = i.select('img')[2].attrs['alt']
        
        movie_grade.append([mv_title, mv_mark, mv_ranking, mv_ud])

movie_grade

    
```

```python
df = pd.DataFrame(movie_grade, columns = ['Movie Title','Movie Point','Ranking Change','Up-Down'])
```
