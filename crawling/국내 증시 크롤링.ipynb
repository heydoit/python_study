```python
#한번에 나타내기
url = "https://finance.naver.com/sise/sise_quant.nhn"
headers = ({
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
})
res = requests.get(url, headers = headers)
soup = bs(res.text, "html.parser")
stock = soup.select('table.type_2  tr')

total = []
for i in stock:
    if len(i.select('td')) == 12:
        no = i.select('td')[0].text
        price = i.select('td')[1].text
        pre = i.select('td')[2].text.replace(',','')
        percentage = i.select('td')[3].text.strip()
        pre_price = i.select('td')[4].text.strip()
        buy = i.select('td')[5].text
        sell = i.select('td')[6].text
        PER = i.select('td')[7].text
        ROE = i.select('td')[8].text
        
        total.append({
            'no' : no,
            'price' : price,
            'pre' : pre,
            'percentage' : percentage,
            'pre price' : pre_price,
            'buy' : buy,
            'sell' : sell,
            'PER' : PER,
            'ROE' : ROE})
df = pd.DataFrame(total)
df
    
```
