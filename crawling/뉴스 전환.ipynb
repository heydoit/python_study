```python
mydate = '20221128'
mypage = 1
headers = ({"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"})
    
total = []

while True:
    url = f'https://news.naver.com/main/list.naver?mode=LS2D&sid2=230&sid1=105&mid=shm&date={mydate}&page={mypage}'
    res = requests.get(url, headers = headers)
    soup = bs(res.text, 'html.parser')
    result = soup.select('.list_body ul li')
    
    if str(mypage) != soup.select_one('div.paging > strong').text.strip():
        print('--> 뉴스목록 {}페이지를 끝으로 크롤링이 완료됐습니다.'.format(soup.select_one('.paging > strong').text.strip()))
        print('-'*80)
        break
        
    #개별 뉴스 조회    
    for count, i in enumerate(result):
        try:
            i.select('dt')[1].select_one('a').text.strip()
        except:
            news_url = i.select('dt')[0].select_one('a').attrs['href']
        else:
            news_url = i.select('dt')[1].select_one('a').attrs['href']
        
        news_res = requests.get(news_url, headers = headers)
        news_soup = bs(news_res.text, 'html.parser')
    
        #뉴스 제목
        try:
            news_soup.select_one('.media_end_head_title h2.media_end_head_headline').text.strip()
        except:
            news_title = news_soup.select_one('div.news_headline h4.title').text.strip()
        else:
            news_title = news_soup.select_one('.media_end_head_title h2.media_end_head_headline').text.strip()
        
        #개별 내용
        
        try:
            news_soup.select_one('div#newsct_article').text.strip()
        except:
            news_content = news_soup.select_one('div#newsEndcontents').text.strip()
        else:
            news_content = news_soup.select_one('div#newsct_article').text.strip()
        
        #작성일지
        try:
            news_soup.select_one('div.media_end_head_info_datestamp_bunch span.media_end_head_info_datestamp_time').text.strip()
        except:
            news_date = news_soup.select_one('div.news_headline div.info').text.strip()
        else:
            news_date = news_soup.select_one('div.media_end_head_info_datestamp_bunch span.media_end_head_info_datestamp_time').text.strip()
        
        total.append([news_date, news_title, news_content])
        
        print('--현재 뉴스목록 {}페이지의 {}번째 개별 뉴스가 로딩됨'.format(soup.select_one('.paging > strong').text.strip(), count+1))
        print('-'*80)
        
    print('#'*80)
    print('===> 현재 뉴스목록 {}페이지의 크롤링이 진행됨'.format(soup.select_one('.paging > strong').text.strip()))
    mypage += 1
```
