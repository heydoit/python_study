```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
browser = webdriver.Chrome()
browser.get('http://www.jungle.co.kr')
```

```python
elm = browser.find_elements(By.CSS_SELECTOR, '.container_main  li')
```

```python
for i in elm:
    print(i.find_elements(By.TAG_NAME, 'a')[0].text)
    print(i.find_elements(By.TAG_NAME, 'a')[1].text)
    print('-'*80)
```
