```python
def musinsa(start, end=None):
    if end == None:
        end = start +1
    else:
        end = end + 1

    import requests
    from bs4 import BeautifulSoup as bs
    import pandas as pd
    
    total = []
    
    for pg in range(start, end):
        url = f"https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page={pg}&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure="
        print(f'{pg} 페이지 크롤링이 완료되었습니다.')
        print()
        print('page',pg, url)
    
        res = requests.get(url)
        soup = bs(res.text, "html.parser")
        result = soup.select('#searchList > li')
    
        for i in result:
            brand = i.select_one('.item_title a').text
            name = i.select_one('.list_info a').attrs['title'].strip()
            if len(i.select_one('.price').text.strip().split()) == 1:
                price = i.select_one('.price').text.strip().split()[0]
            else:
                price = i.select_one('.price').text.strip().split()[1]
            buy_count = i.select_one('.count').text.strip()
            print()
            print('Brand : ', brand)
            print('Name : ', name)
            print('Price : ', price)
            print('Buy count : ', buy_count,'개')
            print('*****************************************')
            total.append({
                'Brand' : brand,
                'Name' : name,
                'Price': price,
                'Buy count':buy_count})
            
    df = pd.DataFrame(total)
    return(df)
        
        
```
