#### 서울 고시원 중 지하시설을 갖춘 고시원의 시설완비 증명발급 현황

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:/csv_file/seoul_gosione(20015_2020)_v1.csv',encoding = 'cp949')
```

```python
data = data.replace(0, np.NaN)
```

```python
data = data.dropna(axis = 0)
```

```python
dataset = data.drop(['연번','업종','지번 주소','도로명 주소','연도','관할소방서'],axis = 1)
data.sort_values(by = '완비증명 발급일', ascending = True)
```

```python
pd.to_datetime(dataset['완비증명 발급일'])
```

```python
dataset['완비증명 발급일'] = pd.to_datetime(dataset['완비증명 발급일'],format='%Y-%m-%d')
```

```python
dataset['완비증명 발급일'] = dataset['완비증명 발급일'].dt.strftime('%Y')
```

```python
dataset
```

```python
dataset_under = dataset[dataset['영업장 층'].str.contains('지하')]
```

```python
dataset_under
```

```python
dataset_under = dataset_under.sort_values(by = '완비증명 발급일', ascending = True)
```

```python
dataset_under
```

```python
dataset_under = dataset_under.drop_duplicates(['영업장 층','상호명'])
```

```python
dataset_under
```

```python
dataset_under['완비증명 발급일'] = dataset_under['완비증명 발급일'].astype(int)
```

```python
dataset_under.info()
```

```python
count90 = dataset_under[(1990 <= dataset_under['완비증명 발급일']) & (dataset_under['완비증명 발급일'] < 2000)]
```

```python
count00 = dataset_under[(2000 <= dataset_under['완비증명 발급일']) & (dataset_under['완비증명 발급일'] < 2010)]
```

```python
count10 = dataset_under[(2010 <= dataset_under['완비증명 발급일']) & (dataset_under['완비증명 발급일'] < 2020)]
```

```python
count20 = dataset_under[(2020 <= dataset_under['완비증명 발급일'])]
```
