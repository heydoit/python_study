```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page=1&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure='
res = requests.get(url)
soup = bs(res.text, 'html.parser')
```

```python
img = soup.select('#searchList > li .lazyload')
```

```python
for i in img:
    print('http:' + i.attrs['data-original'])

```

```python
for idx, i in enumerate(img):
    print('http:' + i.attrs['data-original'])
    img_url = 'http:' + i.attrs['data-original']
    img_res = requests.get(img_url)
    with open('img{}.jpg'.format(idx+1), mode = 'wb') as f:
        f.write(img_res.content)
```
