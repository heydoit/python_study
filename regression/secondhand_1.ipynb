```python
A = pd.DataFrame(data = {'fuel': ['gasoline', 'hybrid', 'LPG', 'hybrid', 'gasoline', 'gasoline', 'LPG', 'gasoline', 'hybrid', 'hybrid'],
                        'trans':['auto', 'manual', 'CVT', 'CVT', 'auto', 'auto', 'manual', 'CVT', 'auto', 'manual']})

A
```
