데이터 정규화 Scalling

```python
import matplotlib.pyplot as plt
import pandas as pd

import sklearn
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
df = pd.read_csv('C:/machine_learning/pyml/data/house_price.csv')
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression() 
```

```python
reg.fit(x, y) 
y_pred = reg.predict(x) 
y_pred
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0) 
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
```

```python
minmax.fit(x_train)
x_train_normed = minmax.transform(x_train)
```

```python
x_test_normed = minmax.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
reg.fit(x_train_normed, y_train)
```

```python
print('기울기: ', reg.coef_)
print('절편: ', reg.intercept_)
```

```python
y_train_pred = reg.predict(x_train_normed)
y_test_pred = reg.predict(x_test_normed)
print('y_train_pred ==> ', y_train_pred)
print('y_test_pred ==> ',y_test_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, y_train_pred)
```

### TEST SET 데이터를 통한 확률적경사하강법 모델링

```python
# SGD 분석모델 객체 생성

from sklearn.linear_model import SGDRegressor

sr = SGDRegressor(max_iter=1000, eta0=0.01, random_state= 0, verbose=1)
```

```python
sr.fit(x_train_normed, y_train)
```

```python
y_train_pred_sgd = sr.predict(x_train_normed) 
y_train_pred_sgd
```

```python
y_test_pred_sgd = sr.predict(x_test_normed) 
y_test_pred_sgd
```

```python
from sklearn.metrics import r2_score

r2_score(y_train, y_train_pred_sgd)
```
