## diamonds form

```python
import pandas as pd
import numpy as np

diamonds = pd.read_csv('./FILE/diamonds.csv')
diamonds.head()
```

## crosstab

```python
pd.crosstab(diamonds['cut'], diamonds['clarity'])
```

```python
pd.crosstab(diamonds['cut'], diamonds['clarity'], normalize=True)
# normarlize == > 비중 알 수 있음
```

```python
pd.crosstab(diamonds['cut'], diamonds['clarity'], normalize=0)
```

* 클러스터링 알고리즘 쓰기 직전 사용하는 것 (참고) - 알면 좋은 것

* normalize = 0 : 행 에 대한 비중(전체 Fair 중 l1이 차지하는 비중...)
* normalize = 1 : 열 에 대한 비중(전체 l1 중 Fair이 차지하는 비중...)

```python
pd.crosstab(diamonds['cut'], diamonds['clarity'], values = diamonds['price'],
            aggfunc = pd.Series.mean)
##가격 평균
```

```python
# 위에 항목이랑 같은 것

diamonds.groupby(['cut','clarity'])['price'].mean().reset_index()
```

```python
dia_agg = diamonds.groupby(['cut','clarity'])['price'].mean().reset_index()
dia_agg.head()
```

* cut과 clarity로 그룹을 지어 price 의 평균값을 보기 위함.
index 번호는 리셋하여 다시 지정함

```python
print(dia_agg.sort_values('price').head()) # 내림차순
print()
print(dia_agg.sort_values('price', ascending=False).head()) # 오름차순
```

```python
dia_agg.sort_values(['cut','clarity'], ascending=[True, False]).head()
```

* 두개를 한번에 정렬할 수 있고 리스트로 묶어 각각 내림차순, 오름차순 지정 가능

### 가장 많은 데이터가 있는 세공 수준과 색상 조합을 순서대로 고르면?

```python
dia = pd.read_csv('./FILE/diamonds.csv')
dia.head()
```

```python
dia_cross = pd.crosstab(dia.cut, dia.color)
# dia_cross.melt().sort_values('value')
```

```python
dia_melt = dia_cross.reset_index().melt(id_vars='cut')
dia_melt.sort_values('value',ascending=False).head(2)
```

### 세공수준별 색상별 가격과 케럿의 평균을 구하고 1캐럿당 가격이 높은 세공수준과 색상 조합을 순서대로 고르면?

```python
dia.head()
```

```python
dia_agg = dia.groupby(['cut','color'])[['price','carat']].mean().reset_index()
dia_agg['ratio'] = dia_agg['price'] / dia_agg['carat']
```

```python
dia_agg.sort_values(['ratio'], ascending=False).head()
```
