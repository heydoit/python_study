```python
import pandas as pd
from scipy.stats import ttest_1samp # 단일 표본 t검정 ( 독립 1 표본)
from scipy.stats import ttest_rel # 대응 표본 t검정
from scipy.stats import ttest_ind # 독립 2 표본
```

```python
df = pd.read_csv('./FILE/iris.csv')
df.head()
```

```python
ttest_1samp(df['Sepal.Length'], popmean=4)
```

* 검정통계량: statistic=27.263680640799215, 
* p 검정: pvalue=8.764592435410748e-60

```python
result = ttest_1samp(df['Sepal.Length'], popmean=4)
round(result[0],3)
```

```python
stat, p = ttest_1samp(df['Sepal.Length'], popmean=4)
print(round(stat, 3))
print(round(p,3))
```

```python
df['Sepal.Length'].mean() # 샘플의 평균
```

```python
stat, p = ttest_1samp(df['Sepal.Length'], popmean=5.843) # 모집단의 평균
print(round(stat, 3))
print(round(p,3))
```

```python
stat, p = ttest_rel(df['Sepal.Length'], df['Sepal.Width'])
print(round(stat, 3))
print(round(p,3)) 

# 귀무가설 기각, 대립가설 채택, 두 집단간 유의미한 차이가 있다
```

```python
df.Species.unique()
```

```python
stat, p = ttest_ind(df.loc[df['Species'] == 'setosa', 'Petal.Length'],
                   df.loc[df['Species'] == 'versicolor', 'Petal.Length'])
print(round(stat, 3))
print(round(p,3))
```

### 수집된 지역의 평균온도는 20도라고 한다. 수집된 데이터를 사용하여 양측 검정을 실시했을 때 p값은?

### 2011.1월 데이터를 대상으로 동 시간대의 causal 과 registerd의 평균 차이 검정 시 검정통계량은?

```python
ttest_rel(bike.loc[bike.datetime == '2011-01', 'casual'],
         bike.loc[bike.datetime == '2011-01', 'registered'])
```

```python
bike['datetime'] = pd.to_datetime(bike.datetime)
bike['year'] = bike['datetime'].dt.year
bike['month'] = bike['datetime'].dt.month
bike_sub = bike.loc[(bike.year == 2011) & (bike.month == 1),]
bike_sub
```

```python
stat, p = ttest_rel(bike_sub.casual, bike_sub.registered)
print(round(abs(stat),4))
print(round(p,4))
```

### 주중과 주말의 registered 평균 검정 시 검정통계량은?

```python
ttest_rel(bike.holiday.mean(), bike.workingday.mean())
```

```python
bike['datetime'] = pd.to_datetime(bike['datetime'])
bike['hol'] = bike['datetime'].dt.weekday
bike['work'] = (bike['hol'] >= 5) + 0 
bike
```

```python
stat, p = ttest_ind(bike.loc[bike['work'] == 1, 'registered'],
                   bike.loc[bike['work'] == 0, 'registered'])
print(round(abs(stat),3))
print(round(p, 3))
```
